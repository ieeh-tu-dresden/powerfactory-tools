# Control Example "Multiple Projects"

In case you want to run the some control actions for multiple projects, you can use the following approach.

It allows you to run the control action for each project sequentially while maintaining a single PowerFactory session. 

This can be more efficient than starting and stopping the PowerFactoryInterface for each project.

## Select an compatible PowerFactory Interface
- Import the PowerFactoryInterface (in this example: that is compatible with PowerFactory in version 2025 SP2)
- Specify additional PowerFactory configuration
- Name the PowerFactory projects you want to work with

```python
%load_ext autoreload
%autoreload 2

import logging

from powerfactory_tools.versions.pf2025 import PowerFactoryInterface
from powerfactory_tools.versions.pf2025.interface import ValidPythonVersion

# PowerFactory configuration
PF_SERVICE_PACK = 2  # mandatory
PF_USER_PROFILE = "TBD"  # mandatory, name of the user profile in the PowerFactory data base
PF_INI_NAME = "PowerFactory"  # optional specification of ini file name to switch to full version (e.g. PowerFactoryFull for file PowerFactoryFull.ini)
PF_PYTHON_VERSION = ValidPythonVersion.VERSION_3_13  # python version of local code environment must match the python version of PowerFactory API

# Consider to use raw strings to avoid misinterpretation of special characters, e.g. r"dir\New Project" or r"dir\1-HV-grid".
PROJECT_NAME_1 = "PowerFactory-Tools"  # may be also full path "dir_name\\project_name"
PROJECT_NAME_2 = "Another-Project"  # may be also full path "dir_name\\project_name"
PROJECT_NAME_3 = "Yet-Another-Project"  # may be also full path "dir_name\\project_name"
```

## Activate multiple projects in sequence
Instead of providing the project name as argument to the PowerFactoryInterface, you can activate the desired project within the context using the `join_project` method.

To ensure that a temporary unit conversion, which is performed by default when a project is opened via PowerFactoryInterface, is reverted properly, you need to use the `release_project` method after completing your operations on the project.

```python
# Configure logging to output to the notebook's standard output
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
logger = logging.getLogger()

with PowerFactoryInterface(
    powerfactory_service_pack=PF_SERVICE_PACK,
    powerfactory_user_profile=PF_USER_PROFILE,
    powerfactory_ini_name=PF_INI_NAME,
    python_version=PF_PYTHON_VERSION,
    logging_level=logging.INFO,
    # log_file_path=full_export_path / pathlib.Path("log_pf_control.txt"),  # noqa: ERA001
) as pfi:
    # Activate project 1
    pfi.join_project(PROJECT_NAME_1)
    # Execute some control actions
    pfi.release_project()

    # Activate project 2
    pfi.join_project(PROJECT_NAME_2)
    # Execute some control actions
    pfi.release_project()

    # Activate project 3, but in this example without the default unit conversion
    pfi.join_project(PROJECT_NAME_3, set_default_units=False)
    # Execute some control actions
    pfi.release_project()
```
